# Fetching Stock Data for Date 28 Jan, 2022 for each Ticker via YahooFinance API

```python
yf.pdr_override()
tickers = si.tickers_sp500()
tickers = [item.replace(".", "-") for item in tickers] # Yahoo Finance uses dashes instead of dots
index_name = '^GSPC' # S&P 500
start_date = datetime.datetime.now() - datetime.timedelta(days=3)
end_date = datetime.datetime.now() - datetime.timedelta(days=2)

empty_dataframe_26 = pd.DataFrame()

for ticker in tickers:
  df = pdr.get_data_yahoo(ticker, start_date, end_date)
  df['Ticker']=ticker;
  df=df.reset_index()
  df=df.set_index('Ticker')
  empty_dataframe_26=empty_dataframe_26.append(df)
```

# Fetching the Close Values for each stock on the previous market day

```python
tickers = si.tickers_sp500()
tickers = [item.replace(".", "-") for item in tickers] # Yahoo Finance uses dashes instead of dots
index_name = '^GSPC' # S&P 500
start_date = datetime.datetime.now() - datetime.timedelta(days=4)
end_date = datetime.datetime.now() - datetime.timedelta(days=3)

empty_list_25=[]
empty_dataframe_25 = pd.DataFrame()

# Find top 30% performing stocks (relative to the S&P 500)
for ticker in tickers:
  df = pdr.get_data_yahoo(ticker, start_date, end_date)
  df=df.reset_index()
  empty_list_25.append(df.loc[0]['Close'])
  df['Ticker']=ticker
  df=df.set_index('Ticker')
  empty_dataframe_25=empty_dataframe_25.append(df)
```
